```python
bin_edges = [0, 1, 2, 3, 4, 5]  # Define bin edges
bin_labels = ['0-2', '2-4', '4-6', '6-8', '8-10']  # Define bin labels
df['Carat Range'] = pd.cut(df['Carat Weight'], bins=bin_edges, labels=bin_labels, include_lowest=True)


plt.figure(figsize=(15, 5))
sales_age = df.groupby(['Carat Range'], as_index=False)['Price'].mean().sort_values(by='Price', ascending=False)

sns.barplot(x = 'Carat Range',y= 'Price' ,data = sales_age)
```
